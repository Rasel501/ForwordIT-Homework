# 2nd (Home Work)

```python
dictionary = {}
for item in df['Category']:
    dictionary[item] = dictionary.get(item, 0) + 1
    
print(dictionary)
```

```python
New_Category = sorted(dictionary.items(), key=lambda x: x[1], reverse=True)

for i in New_Category:
    print(i[0], i[1])
```
